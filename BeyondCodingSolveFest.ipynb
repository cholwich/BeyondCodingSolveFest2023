```python
import numpy as np
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
```

# Generating Linear Data

```python
X1, y1 = make_regression(n_samples=100, n_features=1, noise=20,
                         bias=0.0, random_state=2801)
X1trn = X1[:-20]
X1tst = X1[-20:]
y1trn = y1[:-20]
y1tst = y1[-20:]
s = np.argsort(X1tst, axis=0).flatten()
X1srt = X1tst[s]
```

# Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X1trn, y1trn)
y1prd = model.predict(X1tst)
```

```python
y1srt = model.predict(X1srt)
plt.scatter(X1tst, y1tst)
plt.plot(X1srt, y1srt, "r", ls='--')
rmse = np.sqrt(mean_squared_error(y1tst, y1prd))
r2 = r2_score(y1tst, y1prd)
print(f"RMSE = {rmse:.4f}")
print(f"R2 = {r2:.4f}")
```

# Linear Support Vector Regression

```python
from sklearn.svm import LinearSVR

model = LinearSVR(loss='squared_epsilon_insensitive')
model.fit(X1trn, y1trn)
y1prd = model.predict(X1tst)
```

```python
# Plotting
y1srt = model.predict(X1srt)
plt.scatter(X1tst, y1tst)
plt.plot(X1srt, y1srt, "purple", ls='--')

# Evaluation
rmse = np.sqrt(mean_squared_error(y1tst, y1prd))
r2 = r2_score(y1tst, y1prd)
print(f"RMSE = {rmse:.4f}")
print(f"R2 = {r2:.4f}")
```

# Generate Polynomial Data

```python
np.random.seed(28)
X2 = 2 - 3*np.random.normal(0, 1, 100)
y2 = -5*X2 - 2*(X2**2) + 3*(X2**3) + np.random.normal(-3, 3, 100)
idx = np.arange(100)
np.random.shuffle(idx)
X2 = X2.reshape((-1, 1))
X2trn = X2[idx[:-20]]
X2tst = X2[idx[-20:]]
y2trn = y2[idx[:-20]]
y2tst = y2[idx[-20:]]

s = np.argsort(X2tst, axis=0).flatten()
X2srt = X2tst[s]
plt.scatter(X2trn, y2trn)
```

# Linear Regression

```python
from sklearn.preprocessing import PolynomialFeatures

model = LinearRegression()
model.fit(X2trn, y2trn)

# Plotting
X2srt = X2tst[s]
y2plt = model.predict(X2srt)
plt.scatter(X2tst, y2tst)
plt.plot(X2srt, y2plt, "red", ls='--')

# Evaluation
y2prd = model.predict(X2tst)
rmse = np.sqrt(mean_squared_error(y2tst, y2prd))
r2 = r2_score(y2tst, y2prd)
print(f"RMSE = {rmse:.4f}")
print(f"R2 = {r2:.4f}")
```

# Polynomial Regression

## Degree 3

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=3)
X3 = poly_features.fit_transform(X2)

X3trn = X3[idx[:-20]]
X3tst = X3[idx[-20:]]
y3trn = y2[idx[:-20]]
y3tst = y2[idx[-20:]]

model = LinearRegression()
model.fit(X3trn, y3trn)

# Plotting
X3srt = X3tst[s]
y3plt = model.predict(X3srt)
plt.scatter(X2tst, y2tst)
plt.plot(X2srt, y2plt, "red", ls='--')
plt.plot(X2srt, y3plt, "g", ls='--')

# Evaluation
y3prd = model.predict(X3tst)
rmse = np.sqrt(mean_squared_error(y3tst, y3prd))
r2 = r2_score(y3tst, y3prd)
print(f"RMSE = {rmse:.4f}")
print(f"R2 = {r2:.4f}")
```
